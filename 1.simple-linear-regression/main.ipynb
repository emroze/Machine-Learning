```python
#Simple linear regression data generation
import matplotlib.pyplot as plt
import numpy as np

rng =  np.random.RandomState(42) #Setting seed for random data generation
x = 10*rng.rand(50) #Generate 50 point along x axis 
y = 2*x-1 + rng.rand(50) #Generate value of y and add random value to simulate real data

plt.scatter(x,y) #plot the data
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)

model
```

```python
X = x[:, np.newaxis] #Reshaping data cause scikit-learn require 2D data

print("Previous data shape", x.shape)
print("Previous Data",x)
print("Current data shape", X.shape)
print("Current Data",X)
```

```python
model.fit(X,y)
```

```python
print("Coefficient",model.coef_)
print("Intercept", model.intercept_)
```

```python
#Predicting

x_pred = np.linspace(-1,11) #generate some numbers to feed to the model
x_pred = x_pred[:,np.newaxis]
y_fit = model.predict(x_pred)
```

```python
plt.scatter(x, y) #Ploting main data
plt.plot(x_pred, y_fit) #Showing the line
```
